```python
import pandas as pd
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

# Load and preprocess data
df = pd.read_csv('Auto Sales data.csv')
```

```python

# Regression Features
regression_features = ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'MSRP', 'DAYS_SINCE_LASTORDER']
X_reg = df[regression_features]
y_reg = df['SALES']
```

```python

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=1)
```

```python
# Model training
reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)
y_pred_reg = reg_model.predict(X_test_reg)
```

```python

# Error metrics
print("Regression Metrics:")
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
print("RMSE:", mean_squared_error(y_test_reg, y_pred_reg, squared=False))
print("R2 Score:", r2_score(y_test_reg, y_pred_reg))
```

```python

# -------------------- Classification Model --------------------
# Objective: Predict DEALSIZE category based on order features
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Encode DEALSIZE
df = df[df['DEALSIZE'].notnull()]
X_clf = df[regression_features]
y_clf = df['DEALSIZE']
```

```python

# Train-test split
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=1)
```

```python

# Model training
clf_model = DecisionTreeClassifier()
clf_model.fit(X_train_clf, y_train_clf)
y_pred_clf = clf_model.predict(X_test_clf)
```

```python

# Accuracy and report
print("\nClassification Metrics:")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_clf))
print("Classification Report:\n", classification_report(y_test_clf, y_pred_clf))
```
