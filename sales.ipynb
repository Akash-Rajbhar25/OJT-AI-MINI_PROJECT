```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('Auto Sales data.csv')

# Basic Stats
print(df.describe())
print(df.info())

```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data Types:\n", df.dtypes)
print("Missing Values:\n", df.isnull().sum())
print("Summary Statistics:\n", df.describe(include='all'))
```

```python

# -------------------- Data Cleaning --------------------
# Strip whitespace from column names and string fields
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['SALES'], bins=30, kde=True)
plt.title('Distribution of Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='STATUS')
plt.title('Order Status Distribution')
plt.tight_layout()
plt.show()






```

```python

# -------------------- Bivariate Analysis --------------------
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='PRODUCTLINE', y='SALES', estimator=sum, ci=None)
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='DEALSIZE', y='SALES')
plt.title('Sales by Deal Size')
plt.tight_layout()
plt.show()
```

```python

# -------------------- Time Series Analysis --------------------
daily_sales = df.groupby('ORDERDATE')['SALES'].sum()
daily_sales.plot(figsize=(10,4), title='Daily Sales Over Time')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python

# -------------------- Correlation Analysis --------------------
num_cols = ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'SALES', 'MSRP', 'DAYS_SINCE_LASTORDER']
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python

# -------------------- Grouped Analysis --------------------
sales_by_country = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)
print("Total Sales by Country:\n", sales_by_country)

sales_by_customer = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(10)
print("Top 10 Customers by Sales:\n", sales_by_customer)

avg_days_by_status = df.groupby('STATUS')['DAYS_SINCE_LASTORDER'].mean()
print("Average Days Since Last Order by Status:\n", avg_days_by_status)

```
